Scrape images from dermnetnz.org

```python
import requests
from bs4 import BeautifulSoup
import os

# Function to download an image given a URL
def download_image(url, folder):
    response = requests.get(url, stream=True)
    if response.status_code == 200:
        filename = os.path.join(folder, url.split("/")[-1])
        with open(filename, "wb") as f:
            print(f"Writing file: {filename}")
            for chunk in response.iter_content(1024):
                f.write(chunk)
    else:
        print(f'request failed: {response.status_code}, {response.content}')

# Specify the URL of the website

base_url = "https://dermnetnz.org"
url = f"{base_url}/topics/acne-face-images"
# Send a GET request to the website
response = requests.get(url)

# Create BeautifulSoup object from the response HTML
soup = BeautifulSoup(response.text, "html.parser")

# Find all image tags in the HTML
# image_tags = soup.find_all(".content__main img")
image_tags = soup.find_all("img")

# Create a folder to store the downloaded images
folder = "acne_images"
if not os.path.exists(folder):
    os.makedirs(folder)

# Download each image
for img in image_tags:
    image_path = img["src"]
    if not image_path.startswith("http"):
        image_url = base_url + image_path
        download_image(image_url, folder)
        print(f"Downloaded: {image_url}")

print("Image scraping completed!")
```
